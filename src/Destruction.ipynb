```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
labels = pd.read_csv('../csv/train_labels.csv')
labels.head()
```

```python
values = pd.read_csv('../csv/train_values.csv')
values.head(10).T
```

```python
id_position = values[["building_id", "position"]]
merge_position_destruction = id_position.merge(labels, on="building_id")
damage_by_position = pd.pivot_table(merge_position_destruction, index='damage_grade', columns='position',\
                                    values='building_id', aggfunc="count")
damage_by_position.plot(kind="pie",\
     figsize= (15, 10),\
     rot=45, \
     autopct = "%.2f%%", \
     colors = ["#11bbff", "#2277ff", "#3300ff"], \
     subplots=True)
```
